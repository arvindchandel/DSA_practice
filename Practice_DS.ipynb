```python
# 2 sum problem
def twoSum(nums, target):
    """
    :type nums: List[int]
    :type target: int
    :rtype: List[int]
    """
    num_dict = {}
    for i in range(len(nums)):
        complement = target - nums[i]
        if complement in num_dict:
            return [num_dict[complement], i]
        num_dict[nums[i]] = i
    return []

if __name__=='__main__':
    res = twoSum([2,0,9,1,4,8,7,6],17)
    print(f'res={res}')
```

```python
# Find First and Last Position of Element in Sorted Array
def find_first_and_last_position(nums, target):
    def find_first(nums, target):
        left, right = 0, len(nums) - 1
        first = -1

        while left <= right:
            mid = left + (right - left) // 2

            if nums[mid] == target:
                first = mid
                right = mid - 1
            elif nums[mid] < target:
                left = mid + 1
            else:
                right = mid - 1

        return first

    def find_last(nums, target):
        left, right = 0, len(nums) - 1
        last = -1

        while left <= right:
            mid = left + (right - left) // 2

            if nums[mid] == target:
                last = mid
                left = mid + 1
            elif nums[mid] < target:
                left = mid + 1
            else:
                right = mid - 1

        return last

    first = find_first(nums, target)
    last = find_last(nums, target)

    return [first, last]

# Example usage:
nums = [1, 2, 2, 2, 3, 4, 5, 5, 6]
target = 2
result = find_first_and_last_position(nums, target)
print(result)  # Output: [1, 3]

```

```python
# Longest Substring Without Repeating Characters

def longest_substring(s):
    left, right = 0, 0
    max_length = 0
    chars = set()
    
    while right < len(s):
        if s[right] not in chars:
            chars.add(s[right])
            right += 1
            max_length = max(max_length, right - left)
        else:
            chars.remove(s[left])
            left += 1
    
    return max_length
```

```python
# Container with most water..
def maxArea(height):
    left, right = 0, len(height) - 1
    max_area = 0
    
    while left < right:
        max_area = max(max_area, min(height[left], height[right]) * (right - left))
        
        if height[left] < height[right]:
            left += 1
        else:
            right -= 1
    
    return max_area
```

```python
def remove_element(nums, val):
    # Initialize pointers
    left = 0
    right = len(nums)
    while left < right:
        if nums[left] == val:
            # Move the last element to the current position
            nums[left] = nums[right - 1]
            # Decrease the array size
            right -= 1
        else:
            # Move to the next element
            left += 1

    # Return the count of remaining elements
    return right

nums = [3, 2, 2, 3,4]
val = 3
result = remove_element(nums, val)
print(result)  # Output: 2
```

```python
def majority_element(nums):
    count = 0
    candidate = None
    # Voting process
    for num in nums:
        if count == 0:
            candidate = num
        if num == candidate:
            count += 1
        else:
            count -= 1
    # Verifying the majority element
    count = 0
    for num in nums:
        if num == candidate:
            count += 1
    # Checking if the candidate is the majority element
    if count > len(nums) // 2:
        return candidate
    else:
        raise ValueError("No majority element found in the array.")

nums = [2, 4, 2, 1, 2, 3, 2, 2, 4, 2, 2]
result = majority_element(nums)
print(result)
```

```python
class StringSplitter:
    def is_palindrome(self, string):
        return string == string[::-1]

    def split_into_palindromes(self, s):
        n = len(s)

        for i in range(1, n - 1):
            for j in range(i + 1, n):
                first = s[:i]
                second = s[i:j]
                third = s[j:]

                if self.is_palindrome(first) and self.is_palindrome(second) and self.is_palindrome(third):
                    return True

        return False
```

```python
# Anagram string
# 1st Solution
from collections import Counter

def are_anagrams(str1, str2):
    # Remove spaces and convert to lowercase to make the comparison case-insensitive
    str1 = str1.replace(" ", "").lower()
    str2 = str2.replace(" ", "").lower()
    
    # Check if character counts are equal
    return Counter(str1) == Counter(str2)

# Example usage:
word1 = "listen"
word2 = "silent"
result = are_anagrams(word1, word2)
print(f"{word1} and {word2} are anagrams: {result}")


# 2nd solution
def are_anagrams(str1, str2):
    # Remove spaces and convert to lowercase to make the comparison case-insensitive
    str1 = str1.replace(" ", "").lower()
    str2 = str2.replace(" ", "").lower()
    
    # Check if the sorted strings are equal
    return sorted(str1) == sorted(str2)

# Example usage:
word1 = "listen"
word2 = "silent"
result = are_anagrams(word1, word2)
print(f"{word1} and {word2} are anagrams: {result}")
```

```python
def longest_palindrome(s):
    n = len(s)
    # Initialize a table to store whether substrings are palindromes
    dp = [[False] * n for _ in range(n)]
    
    # All substrings of length 1 are palindromes
    for i in range(n):
        dp[i][i] = True

    max_length = 1
    start = 0

    # Check substrings of length 2 and greater
    for length in range(2, n + 1):
        for i in range(n - length + 1):
            j = i + length - 1
            if length == 2:
                dp[i][j] = (s[i] == s[j])
            else:
                dp[i][j] = (s[i] == s[j]) and dp[i + 1][j - 1]

            if dp[i][j] and length > max_length:
                max_length = length
                start = i

    return s[start:start + max_length]

# Example usage
input_str = "babad"
print(longest_palindrome(input_str))  # Output: "bab"
```

```python
# Longest increasing sub-sequence  LIS --

def length_of_lis(arr):
    n = len(arr)
    if n == 0:
        return 0

    tails = [0] * n
    length = 1
    tails[0] = arr[0]

    for i in range(1, n):
        if arr[i] < tails[0]:
            tails[0] = arr[i]
        elif arr[i] >= tails[length - 1]:
            tails[length] = arr[i]
            length += 1
        else:
            # Use binary search to find the position to insert arr[i]
            left, right = 0, length - 1
            while left < right:
                mid = left + (right - left) // 2
                if tails[mid] < arr[i]:
                    left = mid + 1
                else:
                    right = mid
            tails[left] = arr[i]

    return length

# Example usage:
arr = [10, 22, 9, 33, 21, 50, 41, 60, 80]
lis_length = length_of_lis(arr)
print(f"Length of LIS: {lis_length}")
```

```python
# maximum sub-array solution

def max_subarray(arr):
    if not arr:
        return None  # Handle the case of an empty array

    max_sum = current_sum = arr[0]
    start = end = 0
    temp_start = 0  # Temporary start index for tracking the current subarray

    for i in range(1, len(arr)):
        if arr[i] > current_sum + arr[i]:
            current_sum = arr[i]
            temp_start = i
        else:
            current_sum += arr[i]

        if current_sum > max_sum:
            max_sum = current_sum
            start = temp_start
            end = i

    return max_sum, arr[start:end + 1]

# Example usage:
arr = [-2, 1, -3, 4, -1, 2, 1, -5, 4]
max_sum, subarray = max_subarray(arr)
print(f"Maximum Subarray Sum: {max_sum}")
print(f"Maximum Subarray: {subarray}")
```

```python
# find minimum in rotated sorted array

def find_min_in_rotated_sorted_array(nums):
    if not nums:
        return None

    left, right = 0, len(nums) - 1

    while left < right:
        mid = left + (right - left) // 2

        # If the mid element is greater than the element at the right, the minimum must be in the right half.
        if nums[mid] > nums[right]:
            left = mid + 1
        # If the mid element is less than the element at the right, the minimum must be in the left half.
        elif nums[mid] < nums[right]:
            right = mid
        # If they are equal, it's safe to decrease the 'right' pointer as the duplicate element can be ignored.
        else:
            right -= 1

    return nums[left]

# Example usage:
nums = [4, 5, 6, 7, 0, 1, 2]
min_element = find_min_in_rotated_sorted_array(nums)
print(f"Minimum element in the rotated sorted array: {min_element}")
```

```python
def search_in_rotated_sorted_array(nums, target):
    left, right = 0, len(nums) - 1

    while left <= right:
        mid = left + (right - left) // 2

        if nums[mid] == target:
            return mid

        if nums[left] <= nums[mid]:
            if nums[left] <= target < nums[mid]:
                right = mid - 1
            else:
                left = mid + 1
        else:
            if nums[mid] < target <= nums[right]:
                left = mid + 1
            else:
                right = mid - 1

    return -1  # Target not found

# Example usage:
nums = [6, 7, 0, 1, 2, 4, 5]
target = 5
index = search_in_rotated_sorted_array(nums, target)
print(f"Target {target} found at index {index}")
```

```python
# Minimum Size Subarray Sum problem

def min_subarray_len(target, nums):
    if not nums:
        return 0

    min_len = float('inf')
    current_sum = 0
    left = 0

    for right in range(len(nums)):
        current_sum += nums[right]

        while current_sum >= target:
            min_len = min(min_len, right - left + 1)
            current_sum -= nums[left]
            left += 1

    return min_len if min_len != float('inf') else 0

# Example usage:
nums = [2, 3, 1, 2, 4, 3]
target = 20
min_length = min_subarray_len(target, nums)
print(f"Minimum size of subarray with sum >= {target}: {min_length}")
```

```python
# Kth largest element of stream..
import heapq

class KthLargest:
    def __init__(self, k, nums):
        self.k = k
        self.min_heap = []
        for num in nums:
            self.add(num)

    def add(self, val):
        if len(self.min_heap) < self.k:
            heapq.heappush(self.min_heap, val)
        else:
            if val > self.min_heap[0]:
                heapq.heappop(self.min_heap)
                heapq.heappush(self.min_heap, val)
        return self.min_heap[0]

# Example usage:
k = 3
stream = [4, 5, 8, 2]
kth_largest = KthLargest(k, stream)
print(kth_largest.add(3))  # Output: 4
print(kth_largest.add(5))  # Output: 5
print(kth_largest.add(10)) # Output: 5
print(kth_largest.add(9))  # Output: 8
print(kth_largest.add(4))  # Output: 8
```

```python
import heapq

# O(n + m + K * log(m))  - time complexity of solution

def topKFrequent(nums, k):
    # Create a dictionary to count the frequency of each element
    freq_map = {}
    for num in nums:
        freq_map[num] = freq_map.get(num, 0) + 1

    # Create a max-heap to store the K most frequent elements
    max_heap = [(-freq, num) for num, freq in freq_map.items()]
    heapq.heapify(max_heap)

    # Get the K most frequent elements from the max-heap
    top_k = []
    for _ in range(k):
        top_k.append(heapq.heappop(max_heap)[1])

    return top_k

# Example usage:
nums = [1, 1, 1, 2, 2, 3]
k = 2
print(topKFrequent(nums, k))  # Output: [1, 2]
```

```python
import heapq

# O(K * (M + N) * log(K))
def kSmallestPairs(nums1, nums2, k):
    if not nums1 or not nums2:
        return []

    # Create a min-heap to store pairs (sum, num1, num2)
    min_heap = []
    for num1 in nums1:
        for num2 in nums2:
            heapq.heappush(min_heap, (num1 + num2, num1, num2))

    # Extract the K smallest pairs from the min-heap
    result = []
    for _ in range(min(k, len(min_heap))):
        _, num1, num2 = heapq.heappop(min_heap)
        result.append([num1, num2])

    return result

# Example usage:
nums1 = [1, 7, 11]
nums2 = [2, 4, 6]
k = 3
print(kSmallestPairs(nums1, nums2, k))  # Output: [[1, 2], [1, 4], [1, 6]]
```

```python
# Sub Array Sum equals K:
# The time complexity of this solution is O(n)

def subarraySum(nums, k):
    # Create a dictionary to store cumulative sums and their frequencies
    cumulative_sums = {0: 1}  # Initialize with 0 to count subarrays starting from the beginning
    count = 0
    cum_sum = 0  # Cumulative sum of elements encountered so far

    for num in nums:
        cum_sum += num

        # Check if there exists a cumulative sum 'pre_sum' such that (cum_sum - pre_sum) equals K
        if cum_sum - k in cumulative_sums:
            # below line is important to understand
            count += cumulative_sums[cum_sum - k]

        # Update the frequency of the current cumulative sum
        cumulative_sums[cum_sum] = cumulative_sums.get(cum_sum, 0) + 1

    return count

# Example usage:
nums = [1, 2, 3, 4, 5]
k = 6
print(subarraySum(nums, k))  # Output: 2 (The subarrays [2, 3, 4, 5] and [4, 5] have a sum of 10)
```

```python
# subsets(the power set) in python

def subsets(nums):
    def backtrack(start, current_subset):
        all_subsets.append(current_subset[:])
        print('what is start:', start)
        for i in range(start, len(nums)):
            current_subset.append(nums[i])
            backtrack(i + 1, current_subset)
            current_subset.pop()
    
    all_subsets = []
    backtrack(0, [])
    return all_subsets

# Example usage:
nums = [1, 2, 3]
result = subsets(nums)

for subset in result:
    print(subset)
```

```python
# combination sum problem python

def combinationSum(candidates, target):
    def backtrack(start, target, path):
        if target == 0:
            result.append(path)
            return
        if target < 0:
            return

        for i in range(start, len(candidates)):
            backtrack(i, target - candidates[i], path + [candidates[i])

    result = []
    backtrack(0, target, [])
    return result

# Example usage:
candidates = [2, 3, 6, 7]
target = 7
combinations = combinationSum(candidates, target)
print(combinations)
```
